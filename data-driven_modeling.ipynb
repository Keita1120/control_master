**伝達関数モデルの記述**

まず，制御工学のモデルの基本である伝達関数モデルを記述してみましょう．Pythonでは，関数tf()を利用して伝達関数モデルを記述します．
なお，時間の都合上，伝達関数などの理論面の説明は省略しますので，各自復習してください．


```
num = [2,1]  # 伝達関数の分子多項式の係数 (2s+1)
den = [1,2,3]  # 伝達関数の分母多項式の係数 (s^2+2s+3)

P = mt.tf(num,den)  # 伝達関数モデルの作成 P = num/den = (2s+1)/(s^2+2s+3)
print(P)
```

**1次遅れ系の時間応答**

つづいて，伝達関数の基本である1次遅れ系の時間応答についてみていきます．特に，1次遅れ系にステップ入力を印加したときのステップ応答について調べましょう．復習ですが，1次遅れ系はゲイン$K$と時定数$T$を用いて

$P(s)=\frac{K}{1+Ts}$

で表わされます．そしてステップ応答はPythonのstep関数で計算されます．

```
T = 0.5 # 時定数の設定
K = 1 # ゲインの設定

Tsim = np.arange(0,5,0.01) # シミュレーション時間の定義

P1 = mt.tf([0,K],[T,1])  # 1次遅れ系の定義

y, t = mt.step(P1,Tsim)  # 引数はモデルとシミュレーション時間，返り値は出力応答と時間

# ステップ応答の描画

plt.figure(figsize = (8,6))
plt.plot(t,y,c = 'blue')
plt.grid()
plt.xlabel('t')
plt.ylabel('y')
plt.show()
```

つぎに，時定数$T$とゲイン$K$をさまざまな値に変化させてステップ応答の違いを見てましょう．まず，ゲイン$K$を固定して時定数$T$を変化せてみます．

```
K = 1  # ゲインKの定義
T = [1, 0.5, 0.1]  # 3種類の時定数Tをリストで格納

Tsim = np.arange(0,5,0.01) # シミュレーション時間の定義

colors = ['blue','red','green']  # 3種類のカラーリストの定義

# ステップ応答の描画

plt.figure(figsize = (8,6))
plt.grid()
plt.xlabel('t')
plt.ylabel('y')

for i in range(len(T)):
    P1 = mt.tf([0,K],[T[i],1])  # 1次遅れ系の定義
    y, t = mt.step(P1,Tsim)  # ステップ応答の計算
    plt.plot(t,y,c = colors[i],label = 'T='+str(T[i]))

plt.legend()
plt.show()
```

時定数$T$を小さくすることで，応答が速くなることが確認できました．次にゲイン$K$を変化させてみましょう．

```
K = [1, 2, 3] # 3種類のゲインKをリストで格納
T = 0.5  # 時定数Tを定義

Tsim = np.arange(0,5,0.01) # シミュレーション時間の定義

colors = ['blue','red','green']  # 3種類のカラーリストの定義

# ステップ応答の描画

plt.figure(figsize = (8,6))
plt.grid()
plt.xlabel('t')
plt.ylabel('y')

for i in range(len(K)):
    P1 = mt.tf([0,K[i]],[T,1])  # 1次遅れ系の定義
    y, t = mt.step(P1,Tsim)  # ステップ応答の計算
    plt.plot(t,y,c = colors[i],label = 'K='+str(K[i]))

plt.legend()
plt.show()
```

ゲイン$K$を大きくすることで$y$の定常値が大きくなることが確認できました．

**周波数応答**

つづいて，2次遅れ系に様々な正弦波入力を加えた時の周波数応答についてみていきましょう．Pythonのlsim関数を利用することで，正弦波入力を加えた場合の応答を調べることができます．

```
P2 = mt.tf([2,1],[1,2,3])  # P2 = (2s+1)/(s^2+2s+3)

Tsim = np.arange(0,5,0.01) # シミュレーション時間の定義

colors = ['black','blue','red','green']  # 3種類のカラーリストの定義

freq = [2,5,10,20]  # 加える正弦波入力の周波数のリスト

# 周波数応答の描画

fig, axes = plt.subplots(2,2,figsize = (12,8))
ax = axes.ravel()

for i in range(len(freq)):

    u = np.sin(freq[i]*Tsim)  # 正弦波入力の定義
    y, t, x0 = mt.lsim(P2,u,Tsim,0) # 正弦波応答の計算

    ax[i].plot(t,u,c = colors[i],label = 'u')  # 正弦波入力のプロット
    ax[i].plot(t,y,c = colors[i],linestyle = '--',label = 'y')  # 正弦波応答のプロット
    ax[i].grid()
    ax[i].legend()
    ax[i].set_title('omega='+str(freq[i]))   
    ax[i].set_xlabel('u')
    ax[i].set_ylabel('y')

fig.tight_layout()  # グラフが重ならないようにする
```

これらの結果より，周波数が低いときに入力と出力の振幅比(ゲイン)が大きくなり，周波数が高いときにゲインが小さくなることがわかります．
同様に，周波数が高くなるにつれ，位相が遅れていることが確認できます．

つづいて，2次遅れ系のボード線図を確認してみましょう．ボード線図はゲイン特性と位相特性を描画した図です．Pythonのbode関数でボード線図を計算することができます．

```
fig, axes = plt.subplots(2,1,figsize = (12,8))

gain, phase, w = mt.bode(P2, mt.logspace(-2,2)) # 各周波数におけるゲイン，周波数を計算

axes[0].semilogx(w, 20*np.log10(gain))  # ゲイン線図の描画
axes[1].semilogx(w, phase*180/np.pi)  # 位相線図の描画

fig.show()
```

**システムの安定性**




つづいて，システムの安定性について考えていきましょう．特に安定性の中でも最も基本的な入出力安定(BIBO安定)についてみていきます．
入出力安定とは，有界な入力信号が加わった時に，出力信号も必ず有界になることをいいます．ちなみに，有界な信号とはすべての時刻において無限大に発散しない信号のことをいいます．入出力安定性のイメージは下図のようになります．

BIBO.jpg

そして，この入出力安定性は伝達関数の極を調べることで確かめることができます．
極とは伝達関数における分母多項式の根(=0になるsの値)のことを指します．
Pythonでは，伝達関数の極を調べるためにpole関数を使います．

```
P1 = mt.tf([0,2],[1,3,2])  
P2 = mt.tf([0,1],[1,-5,1])  

print('P1',P1)
print('P2',P2) 

p1 = mt.pole(P1)
p2 = mt.pole(P2)

print('pole of P1',p1)
print('pole of P2',p2) 
```

システムの入出力安定性のための条件は，伝達関数のすべての極の実部が負であることです．すなわち，伝達関数の極が複素平面の左半面内にあれば，システムは入出力安定となります．では，システムP1とP2の安定性について調べましょう．

```
Tsim = np.arange(0,10,0.1) # シミュレーション時間の定義

y1, t = mt.step(P1,Tsim)  # P1のステップ応答
y2, t = mt.step(P2,Tsim)  # P2のステップ応答

# ステップ応答の描画

fig, axes = plt.subplots(2,1,figsize = (12,8))

axes[0].plot(t,y1,c = 'blue')
axes[0].grid()
axes[0].set_xlabel('t')
axes[0].set_ylabel('y')

axes[1].plot(t,y2,c = 'red')
axes[1].grid()
axes[1].set_xlabel('t')
axes[1].set_ylabel('y')

fig.show()
```

不安定システムP2では，ステップ応答が発散していることが確認できました．このように，伝達関数の極と安定性には深い関係性があることを覚えといてください．

#静的モデリング

ここでは，まずモデリングの基本である静的モデリングについて学びます．静的であるとは，***ある時刻における出力がその時刻における入力と一対一対応するという意味です．***
すなわち，静的モデルとは**時刻によらずに，ある入力に対して必ず同じ出力を返すモデルのことです．**
現時刻$k$の入力を$u(k)$，出力を$y(k)$としたとき
静的モデルの入出力関係は

$y(k)=f(u(k))$

で記述することができます．

**データの準備**

まず，今回説明する静的モデリングで扱う入出力データを定義します．

```
# モジュールの読み込み

import numpy as np
import matplotlib.pyplot as plt
```

```
u = np.linspace(-2,2,100) # -2から2までの100個の値
print(u)
```

```
y_sin = np.sin(u) 
y = y_sin + 0.1 * np.random.randn(100)  # yの値はsinカーブに乱数が加わったもの
```

```
# データの可視化

plt.figure(figsize=(12,8))
plt.scatter(u,y, marker = 'o', c = 'red',label = 'y')  # データyのプロット
plt.plot(u, y_sin, c = "blue", label = 'sin(u)') # 正弦波のプロット
plt.grid()
plt.legend()
plt.show()
```

**線形モデル $y=au$によってデータを回帰**

まず，最も基本的な静的モデルとして，以下のような線形モデル

$y=au$

を考えます．

様々な傾き$a$によって，入出力データをこの線形モデルにフィッティング（回帰）してみましょう

```
# a = 0.1,1,-0.05
a1 = 1
a2 = 0.5
a3 = -0.5

# それぞれの線形モデルによるyの出力値（予測値）
y1 = a1*u 
y2 = a2*u
y3 = a3*u

# データの可視化

plt.figure(figsize=(12,8))
plt.scatter(u,y, marker = 'o', c = 'red',label = 'data')  # データのプロット
plt.plot(u, y1, c = "blue", label = 'a=0.1')  # y1のプロット
plt.plot(u, y2, c = "green", linestyle = '--', label = 'a=0.05') # y2のプロット
plt.plot(u, y3, c = "red", linestyle = ':', label = 'a=-0.05')  # y3のプロット
plt.xlabel('u')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.show()
```

傾き$a$を変化させることで，モデル$y=au$の入出力関係が変化するのが確認できたと思います．ただ，どの傾き$a$が最もデータの特性を反映しているかはわかりません．そこで，データに対してどれくらいあてはめができているかという尺度を設けることによって，データのフィッティング度を定量化します．その尺度としてよく用いられるものが**二乗誤差**です．

二乗誤差は
$J(a) = \sum_{i=1}^{N}\left({y_i - au_i}\right)^2$
によって定義される値です．
この二乗誤差$J(a)$を最小化することで$a$を決定する方法を**最小二乗法**といいます．
最小二乗法のイメージは下図のようになります．

**最小二乗法による係数$a$の決定**

最小二乗法を行なうためのツールとして今回はCVXpyと呼ばれる最適化ツールを用います．CVXpyは最小化したい評価関数や制約条件（今回はなし）を与えると，自動で最適化を行なってくれる便利なツールです．

```
import cvxpy   # cvxpyのインポート
```

```
a_val = cvxpy.Variable(1)  # 求めたいパラメータの定義

objective = cvxpy.Minimize(cvxpy.sum_squares(a_val * u - y))  # コスト関数
constraints = []  # 今回は制約条件なし
prob = cvxpy.Problem(objective, constraints) # 最適化問題を定義

result = prob.solve() # 最適化問題を求解

print("optimal parameter:", a_val.value) # メソッドvalueで値を取得
```

**最小二乗法による係数$a$の決定**

多項式回帰の場合も同様に，CVXpyを使ってパラメータ$\left(a_1,a_2,a_3,b\right)$を求めてみましょう．

今回の場合，コスト関数は

$J(a_1,a_2,a_3,b) = \sum_{i=1}^{N}\{y_i-\left(a_1u_i+a_2u_i^2+a_3u_i^3+b\right)\}^2$

になります．一見複雑そうに見えますが，このコスト関数は求めたいパラメータ$\left(a_1,a_2,a_3,b\right)$に関しては２次（凸）な関数なので求めるのはそう難しくはありません．

```
# 求めたいパラメータの定義

a1_val = cvxpy.Variable()  
a2_val = cvxpy.Variable()
a3_val = cvxpy.Variable()
b_val = cvxpy.Variable()


objective = cvxpy.Minimize(cvxpy.sum_squares(a1_val * u + a2_val * u**2 + a3_val * u**3 + b_val - y))  # コスト関数
constraints = []  # 今回も制約条件なし
prob = cvxpy.Problem(objective, constraints) # 最適化問題を定義

result = prob.solve() # 最適化問題を求解

# メソッドvalueで値を取得
print("optimal parameter(a1):", a1_val.value) 
print("optimal parameter(a2):", a2_val.value)
print("optimal parameter(a3):", a3_val.value) 
print("optimal parameter(b):", b_val.value)  
```

```
a1_ls = a1_val.value
a2_ls = a2_val.value
a3_ls = a3_val.value
b_ls = b_val.value

y_multi_ls = a1_ls * u + a2_ls * u**2 + a3_ls * u**3 + b_ls

# データの可視化

plt.figure(figsize=(12,8))
plt.scatter(u,y, marker = 'o', c = 'red',label = 'data')
plt.plot(u, y_multi_ls, c = "blue", label = 'least_square')
plt.xlabel('u')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.show()
```

このように多項式モデルを導入することで，正弦波の関係を再現することができました．

#動的モデリング

ここでは，動的モデリングについて紹介します．動的モデルとは，**現在の出力に現在の入力のみならず，過去の入力も関係するモデルを指します．**
すなわち時刻$k$における入力を$u(k)$，出力を$y(k)$としたとき，動的モデルの入出力の関係は

$
a_0 y(k)+a_1 y(k-1)+a_2 y(k-2)+\cdots+a_n y(k-n)
=b_0 u(k)+b_1 u(k-1)+b_2 u(k-2)+\cdots+b_mu(k-m)
$

のように記述することができます．

**データの準備**

まず，今回モデリングに用いる入出力の時系列データを読み込みましょう．

```
t = np.linspace(0,100,100)  # 時刻k=0から時刻k=100まで
```

```
udata = np.loadtxt('input_data.csv', delimiter=',')  # 入力データの読み込み
ydata = np.loadtxt('output_data.csv', delimiter=',')  # 出力データの読み込み
```

つづいて，読み込んだ入出力データを可視化してみましょう．

```
plt.figure(figsize=(12,8))
plt.plot(t,udata,c='red',label='input')
plt.plot(t,ydata,c='blue',label='output')
plt.grid()
plt.legend()
plt.show()
```

図のように入出力データは，$1$秒ごとにサンプリングした$100$秒までの時系列データとなっています．





**動的モデル**

今回は，入出力データの関係を記述する動的モデルとして

$
y(k)= a_1 y(k-1)+a_2 y(k-2)+b u(k-1)
$

を仮定します．
入出力データ$\{u(0),\cdots,u(N-1)\},\{y(0),\cdots,y(N)\}$を用いて，動的モデルの係数$a_1,a_2,b$を求めましょう．



**最小二乗法**

まず，入出力の関係を時系列順に並べると

$\begin{array}{llll}
 y(2)&=a_1 y(1)+a_2 y(0)+bu(1) \\
 y(3)&=a_1 y(2)+a_2 y(1)+bu(2) \\
 &\vdots\\
 y(N)&=a_1 y(N-1)+a_2 y(N-2)+bu(N-1) 
 \end{array}$

となります．
静的モデリングと同様に，この左辺と右辺の差を最小化する係数$a_1,a_2,b$を決定しましょう．そのために，ここでも最小二乗法を利用します．

最小二乗法によって最小化するコスト関数は

$\sum_{k=2}^{N}\{y(k)- \left(a_1y(k-1)+a_2y(k-2)+bu(k-1)\right)\}^2$

となります．

このコスト関数をそれぞれパラメータ$\left(a_1,a_2,b\right)$に対して最小化することによって，最もデータにフィッティングする$\left(a_1,a_2,b\right)$を求めることができます．



```
import cvxpy   # cvxpyのインポート
```

```
# 求めたいパラメータの定義 
a1_val = cvxpy.Variable(1) 
a2_val = cvxpy.Variable(1) 
b_val = cvxpy.Variable(1) 

objective = cvxpy.Minimize(cvxpy.sum_squares( ydata[2:] - (a1_val * ydata[1:-1] + a2_val* ydata[:-2] + b_val * udata[1:-1]) ) ) # コスト関数 
# ydata[:-2]は時刻k=0からk=N-2までの出力データ
# ydata[1:-1]は時刻k=1からk=N-1までの出力データ
# ydata[2:]は時刻k=2からk=Nまでの出力データ
# udata[1:-1]は時刻k=1からk=N-1までの入力データ

constraints = [] # 今回は制約条件なし 
prob = cvxpy.Problem(objective, constraints) # 最適化問題を定義

result = prob.solve() # 最適化問題を求解

# メソッドvalueで値を取得
print("a_1:", a1_val.value) 
print("a_2:", a2_val.value)
print("b:", b_val.value)
```

求めたパラメータ$\left(a_1,a_2,b\right)$を用いて，出力データを予測しましょう．ここでは，毎時刻入力データ$\{u(0),\cdots,u(N-1)\}$を動的モデルに印加することにより出力データを再現します．

```
a_1 = a1_val.value
a_2 = a2_val.value
b = b_val.value

y = list(ydata[0:2]) # y(0)とy(1)をリストに格納

y_pre2 = y[0] # 2時刻前の出力
y_pre1 = y[1] # 1時刻前の出力

for ui in udata[1:-1]:

    # y_nowは現時刻の出力y(k)
      
    y_now = a_1 * y_pre1 + a_2 * y_pre2 + b * ui  # 動的モデルによるy(k)の更新
    
    y.append(y_now) # 現時刻の出力y_nowをリストに追加

    y_pre1 = y_now # y(k-1)⇒y(k)に更新
    y_pre2 = y_pre2  #y(k-2)⇒y(k-1)に更新

plt.figure(figsize=(12,8))
plt.plot(t,y,c='blue',label='predict')
plt.plot(t,ydata,c='red',linestyle = '--',label='data')
plt.xlabel('t')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.show()
```

上図の結果より，出力の時系列特性を再現できていることが確認できました．
以上が動的モデリングとなります．動的モデリングは制御工学の基本となりますので，静的モデリングとの違いを意識して，しっかり理解しましょう．

#応用：制約付きモデリング

最後に制約付きモデリングについて学びましょう．今までのモデリングでは，最小二乗法を利用してモデルのパラメータ（係数）を決定していました．その際，無制約のもとコスト関数を最小化していましたが，あらかじめパラメータに関して事前情報（例えば，パラメータの値が正の定数）などが存在する場合，その事前情報を制約としてコスト関数を最小化することでより精度の高いモデリングが可能となります．このように，制約条件を考慮したモデリング法を制約付きモデリングと呼びます．

具体例として，線形モデル

$y=au$

による静的モデリングを考えましょう．
入出力のデータセット$\{u_i,y_i\}$が得られているとして，最小二乗法を適用するとします．このとき，係数$a$は最適化問題

$\min J(a), \quad J(a) = \sum_{i=1}^{N}\left({y_i - au_i}\right)^2$

を求解することで得られます．
一方で，$a\geq0$といった事前情報がある場合，$a\geq0$という制約条件のもと，$J(a)$を最小化する必要があります．この問題は，制約付き最適化問題として

$\min J(a),\quad \mathrm{subject.to}\quad a\geq0$

のように表わすことができます．このイメージは下図で表わすことができます．



スライド1.JPG

ここでは，制約付き最適化問題の例として，先ほど考えた動的モデリングの例を考えましょう．ただし，先ほどと違う点として，動的モデルの係数に制約条件

$a_1\leq-0.1,\quad  a_2\geq0.1,\quad b\geq0.1$

が存在するとします．



```
# 求めたいパラメータの定義 
a1_val = cvxpy.Variable(1) 
a2_val = cvxpy.Variable(1) 
b_val = cvxpy.Variable(1) 

objective = cvxpy.Minimize(cvxpy.sum_squares( ydata[2:] - (a1_val * ydata[1:-1] + a2_val* ydata[:-2] + b_val * udata[1:-1]) ) ) # コスト関数 
# ydata[:-2]は時刻k=0からk=N-2までの出力データ
# ydata[1:-1]は時刻k=1からk=N-1までの出力データ
# ydata[2:]は時刻k=2からk=Nまでの出力データ
# udata[1:-1]は時刻k=1からk=N-1までの入力データ

constraints = [a1_val<=-0.1, a2_val>=0.1, b_val>=0.1]  # 制約条件として a1<=-0.1, a2>=0.1, b>=0.1 が与えられているとする

prob = cvxpy.Problem(objective, constraints) # 最適化問題を定義

result = prob.solve() # 最適化問題を求解

# メソッドvalueで値を取得
print("a_1:", a1_val.value) 
print("a_2:", a2_val.value)
print("b:", b_val.value)
```

上の結果より，最適解が制約条件を満足していることがわかります．

最後に求めたパラメータ$\left(a_1,a_2,b\right)$を用いて，出力データを予測しましょう．

```
a_1 = a1_val.value
a_2 = a2_val.value
b = b_val.value

y = list(ydata[0:2]) # y(0)とy(1)をリストに格納

y_pre2 = y[0] # 2時刻前の出力
y_pre1 = y[1] # 1時刻前の出力

for ui in udata[1:-1]:

    # y_nowは現時刻の出力y(k)
      
    y_now = a_1 * y_pre1 + a_2 * y_pre2 + b * ui  # 動的モデルによるy(k)の更新
    
    y.append(y_now) # 現時刻の出力y_nowをリストに追加

    y_pre1 = y_now # y(k-1)⇒y(k)に更新
    y_pre2 = y_pre2  #y(k-2)⇒y(k-1)に更新

plt.figure(figsize=(12,8))
plt.plot(t,y,c='blue',label='predict')
plt.plot(t,ydata,c='red',linestyle = '--',label='data')
plt.xlabel('t')
plt.ylabel('y')
plt.grid()
plt.legend()
plt.show()
```

 \vdots\\
 x(N) = A^{N}x(0)+A^{N-1}Bu(0)+A^{N-2}Bu(1)+\cdots+ABu(N-2)+Bu(N-1)
$
